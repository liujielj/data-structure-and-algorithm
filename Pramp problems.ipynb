### Bracket Match

A string of brackets is considered correctly matched if every opening bracket in the string can be paired up with a later closing bracket, and vice versa. For instance, “(())()” is correctly matched, whereas “)(“ and “((” aren’t. For instance, “((” could become correctly matched by adding two closing brackets at the end, so you’d return 2.

Given a string that consists of brackets, write a function bracketMatch that takes a bracket string as an input and returns the minimum number of brackets you’d need to add to the input in order to make it correctly matched.

Explain the correctness of your code, and analyze its time and space complexities.

Examples:

input:  text = “(()”
output: 1

input:  text = “(())”
output: 0

input:  text = “())(”
output: 2


```python
# use a stack, from start of the string, initalize the stack by push 1 item
# whenever we push an item, check the top of the stack
# if it's a pair, then we can remove them
# the number of brackets need to add to make it match is the length of the stack
# time complexity: O(n), space complexity: O(n)
def bracket_match(s):
    if not s:
        return 0
    stack = []
    for bracket in s:
        if stack and (bracket == ')' and stack[-1] == '('):
            stack.pop()
        else:
            stack.append(bracket)
    return len(stack)

# test
s = ')()()()((('
bracket_match(s)
```

```python
# another way without stack
# since there is only one type of brakets, so just use a counter number is enough
# keep a counter number, iterate the string s
# when we see left bracket, increment 1; when we see right bracket, minus 1
# when the counter number smaller than 0 (-1), that means we need to compensate it with a left bracket
# result +1 and reset the counter to 0
# final result is the result plus the counter

def bracket_match(s):
    counter = 0
    result = 0
    for bracket in s:
        if bracket == '(':
            counter += 1
        else:
            counter -= 1
        if counter == -1:
            result += 1
            counter = 0
    result += counter
    return result
# test
s = ')()()()((('
```

### Award Budget Cuts


The awards committee of your alma mater (i.e. your college/university) asked for your assistance with a budget allocation problem they’re facing. Originally, the committee planned to give N research grants this year. However, due to spending cutbacks, the budget was reduced to newBudget dollars and now they need to reallocate the grants. The committee made a decision that they’d like to impact as few grant recipients as possible by applying a maximum cap on all grants. Every grant initially planned to be higher than cap will now be exactly cap dollars. Grants less or equal to cap, obviously, won’t be impacted.

Given an array grantsArray of the original grants and the reduced budget newBudget, write a function findGrantsCap that finds in the most efficient manner a cap such that the least number of recipients is impacted and that the new budget constraint is met (i.e. sum of the N reallocated grants equals to newBudget).

Analyze the time and space complexities of your solution.

Example:

input:  grantsArray = [2, 100, 50, 120, 1000], newBudget = 190

output: 47 # and given this cap the new grants array would be
           # [2, 47, 47, 47, 47]. Notice that the sum of the
           # new grants is indeed 190
Constraints:

[time limit] 5000ms

[input] array.double grantsArray

0 ≤ grantsArray.length ≤ 20
0 ≤ grantsArray[i]
[input] double newBudget

[output] double

```python
# solution during practice
def find_grants_cap(grantsArray, newBudget):
  grantsArray.sort()
  total = sum(grantsArray)
  if newBudget >= total:
    return grantsArray[-1]
  # decide how many grants need to be capped
  i = len(grantsArray) - 1
  while total > newBudget and i >= 0:
    if i == 0:
      cap = 0
      total -= (len(grantsArray) - i) * grantsArray[i]
      i -= 1
    else:
      cap = grantsArray[i - 1]
      total -= (len(grantsArray) - i) * (grantsArray[i] - grantsArray[i - 1])
      i -= 1
  # cap = a[i]
  total = sum(grantsArray[:i + 1])
  cap = (newBudget - total) / float(len(grantsArray) - i - 1)
  return cap


# test
print(find_grants_cap([2, 4], 3))
print(find_grants_cap([2, 4, 6], 3))
print(find_grants_cap([2,100,50,120,167], 400))
print(find_grants_cap([2,100,50,120,1000], 190))
print(find_grants_cap([21,100,50,120,130,110], 140))
print(find_grants_cap([210,200,150,193,130,110,209,342,117], 1530))
```

```python
# solution after review
def find_grants_cap(grantsArray, newBudget):
    if not grantsArray or newBudget < 0:
        return -1

    grantsArray.sort(reverse = True)
    n = len(grantsArray)
    grantsArray.append(0)
    total = sum(grantsArray)

    if total <= newBudget:
        return grantsArray[0]

    for i in range(n):
        diff = grantsArray[i] - grantsArray[i + 1]
        total -= (i + 1) * diff
        if total <= newBudget:
            break
    # 0th to ith grants need to be capped
    cap = (newBudget - sum(grantsArray[i + 1:])) / float(i + 1)
    return cap

# test
print(find_grants_cap([2, 4], 3))
print(find_grants_cap([2, 4, 6], 3))
print(find_grants_cap([2,100,50,120,167], 400))
print(find_grants_cap([2,100,50,120,1000], 190))
print(find_grants_cap([21,100,50,120,130,110], 140))
print(find_grants_cap([210,200,150,193,130,110,209,342,117], 1530))
```

### Find minimum adjustment cost of an array

We need to minimize the adjustment cost, that is the sum of differences between new and old values. We basically need to minimize ∑|A[i] – Anew[i]| where 0 ≤ i ≤ n-1, n is size of A[] and Anew[] is the array with adjacent difference less that or equal to target.
Assume all elements of the array is less than constant M = 100.
Examples: 
 

Input: arr = [1, 3, 0, 3], target = 1
Output: Minimum adjustment cost is 3
Explanation: One of the possible solutions 
is [2, 3, 2, 3]

Input: arr = [2, 3, 2, 3], target = 1
Output: Minimum adjustment cost is 0
Explanation:  All adjacent elements in the input 
array are already less than equal to given target

             25, 60, 49, 47], target = 10
Output: Minimum adjustment cost is 75
Explanation: One of the possible solutions is 
[55, 62, 52, 49, 39, 29, 30, 40, 49, 47]

```python
# Matrix Spiral Copy
# use recursion to solve this problem
# base case: if the matrix has 0 row or column, return
# recursive rule: 
#   spiral copy the outer ring
#   recursive the the inner matrix
# time complexity: O(m*n)
# space complexity: O(min(m, n))

def spiral_copy(mat):
    result = []
    helper(mat, 0, result)
    return result

def helper(mat, offset, result):
    # base case
    m = len(mat)
    n = len(mat[0])
    if 2 * offset >= m or 2 * offset >= n:
        return
    if 2 * offset == m - 1 and 2 * offset == n - 1:
        result.append(mat[offset][offset])
        return
    # recursive rule
    for i in range(offset, n - 1 - offset):
        result.append(mat[offset][i])
    for j in range(offset, m - 1 - offset):
        result.append(mat[j][n - 1 - offset])
    for i in range(n - 1 - offset, offset, -1):
        result.append(mat[m - 1 - offset][i])
    for j in range(m - 1 - offset, offset, -1):
        result.append(mat[j][offset])
    helper(mat, offset + 1, result)

# test
inputMatrix  = [ [1, 2, 3] 
                ,[6, 5, 4]]
print(spiral_copy(inputMatrix))

inputMatrix  = [ [1, 2, 3] 
                ,[8, 9, 4]
                ,[7, 6, 5]]
print(spiral_copy(inputMatrix))
```

```python
# input: shiftArr = [9, 12, 17, 2, 4, 5], num = 2 
# output: 3
def shifted_arr_search(shiftArr, num):
    # find the pivot index
    pivot = find_pivot(shiftArr)
    if pivot == 0 or num < shiftArr[0]:
        return binary_search(shiftArr, pivot, len(shiftArr) - 1, num)
    else:
        return binary_search(shiftArr, 0, pivot - 1, num)
    
def find_pivot(shiftArr):
    if not shiftArr or len(shiftArr) == 1:
        return 0
    l, r = 0, len(shiftArr) - 1
    while l <= r:
        mid = l + (r - l) // 2
        if mid == 0 or shiftArr[mid] < shiftArr[mid - 1]:
            return mid
        elif shiftArr[mid] < shiftArr[0]:
            r = mid - 1
        else:
            l = mid + 1
    return 0

def binary_search(arr, l, r, num):
    while l <= r:
        mid = l + (r - l) // 2
        if arr[mid] == num:
            return mid
        elif arr[mid] > num:
            r = mid - 1
        else:
            l = mid + 1
    return -1

  
shiftArr = [2]
num = 2
print(shifted_arr_search(shiftArr, num))

shiftArr = [9, 12, 17, 2, 4, 5]
num = 2
print(shifted_arr_search(shiftArr, num))
```
