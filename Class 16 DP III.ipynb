```
# Q0
# largest sum of subarray
# [1, 2, 3, -1, -10, 3]

def largest_sum(a):
  # Base case
  if len(a) == 0:
    return 0
  # Induction case
  # M[i] = the largest sum of subarray ending at a[i]
  curr = 0
  global_max = curr
  for i in range(1, len(a)):
    curr = max(curr + a[i], a[i])
    global_max = max(global_max, curr)
  return global_max

a = [1, 2, 3, -11, 10, -3]
largest_sum(a)
```

```
def rev(num):
    result = 0
    while num:
        result = result * 10 + num % 10
        num = num // 10
    return result

# test
rev(123)
```
