### Common Questions:

#### Given a number x, how to set x's k-th bit to 1?

```
# to set x's k-th bit to 1, we can bitwise or the bit with 1
# set a setter number with only 1 at kth bit, and bitwise or with x
# time complexity: O(1)

def setkth1(x, k):
    return x | 1 << k

# test
setkth1(8, 1)
```

#### Given a number x, how to set x's k-th bit to 0?

```
# to set x's kth bit to 0, we can bitwise and the bit with 0
# set a number with only 0 at kth bit, and bitwise and with x
# time complexity O(1)

def setkth0(x, k):
    return x & ~(1 << k)

# test
setkth0(7, 1)
```

```
# check if x can be divide by 2 without remaining until 0
def checkPower(x):
    if x <= 0:
        return False
    while x > 0:
        if x != 1 and x % 2 != 0:
            return False
        x = x // 2
    return True

# test
checkPower(128)
```

```
# bitwise method
# check if the sum of all bits is 1
def checkPower(x):
    if x <= 0:
        return False
    sum = 0
    while x > 0:
        sum += x & 1
        x >>= 1
    return sum == 1

# test
checkPower(3)
```

```
# bitwise method
# if x is a power of 2, then there is only one of 1 digit bitwise
# and x - 1 has all 1 digit, and length is 1 less than x
# so x & (x - 1) should be 0
def checkPower(x):
    if x <= 0:
        return False
    return x & (x - 1) == 0

# test
checkPower(128)
```

### Question 1: determine whether a number x is a power of 2 (x==2^n) n > 0

```
# non bit opeartion way
def checkPower(x):
  if x <= 1:
    return False

  while x % 2 == 0:
    x //= 2
  return True if x == 1 else False

checkPower(0)
```

```
# bit operation way
def checkPower2(x):
  if x <= 1:
    return False
  
  return x & (x - 1) == 0

checkPower(8)
```

```
# bit operation way
def checkPower2(x):
  if x <= 1:
    return False

  count = 0
  for i in range(64):
    count += (x >> i & 1)
    if count > 1:
      return False
  return True

checkPower(7)
```

### Question 2: How to determine the number of bits that are different between two positive integers?
int x = 0101<br>
int y = 0111<br>
return 1

```
def numOfBits(x, y):
  z = x ^ y
  count = 0
  for i in range(64):
    count += z >> i & 1
  return count

numOfBits(-1, -2)
```

```
def numOfBits(x, y):
  z = x ^ y
  count = 0
  if z < 0:
    z = -z - 1
    count = 1
  while z != 0:
    count += z & 1
    z = z >> 1
  return count

numOfBits(1, -2)
```

### Question 4: determine whether a word contains all letters that are unique (no duplicate letters in the word). 
e.g., input = student, output = false

```
# hashset way
def checkContain(word):
  if not word:
    return True

  checked = set()
  for char in word:
    if char in checked:
      return False
    checked.add(char)
  return True

checkContain('stuS')
```

```
# bit vector
def checkContain2(word):
  if not word:
    return True
  
  checkList = [0] * 8
  for char in word:
    row = ord(char) // 32
    col = ord(char) % 32
    if checkList[row] >> col & 1:
      return False
    checkList[row] = checkList[row] | 1 << col
  return True

checkContain('stus')
```

### Question 5: How to reverse all bits of a number?
e.g., 1010xxxxx0010 → 0100xxxxx0101

```
import sys
def bitReverse(num):
  n = sys.getsizeof(num) * 8
  i, j = 0, n - 1
  while i < j:
    if num >> i & 1 != num >> j & 1:
      num = num ^ (1 << i | 1 << j)
    i += 1
    j -= 1
  return num

bitReverse(15)
```

### Question 6: Given a number x, how to get the hexadecimal representation of the number in string type?
e.g., 29 → "0x1D" 

```
def convertHex(num):
  digits = "0123456789ABCDEF"
  res = []
  while num > 0:
    rem = num % 16
    res.append(digits[rem])
    num = num // 16
  res = res + ['x', '0']
  res.reverse()
  return ''.join(res)

convertHex(17)
```

```
def convertHex(num):
  digits = "0123456789ABCDEF"
  res = ""
  while num > 0:
    rem = num % 16
    res += digits[rem]
    num = num // 16
  res = res + "x0"
  return res[::-1]

convertHex(17)
```
