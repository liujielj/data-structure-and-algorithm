```
# Heapify
# A = [1, 2, 3, 4, 5]
def heapifyTopDown(A, i):
  if i >= len(A):
    return
  left = 2 * i + 1
  right = 2 * i + 2
  idx_max = i
  if left < len(A) and A[left] > A[idx_max]:
    idx_max = left
  if right < len(A) and A[right] > A[idx_max]:
    idx_max = right
  if i != idx_max:
    A[i], A[idx_max] = A[idx_max], A[i]
    heapifyTopDown(A, idx_max)
  
A = [1, 2, 3, 4, 5]
heapifyTopDown(A, 0)
A
```

```
# Build a max-heap
def buildMaxHeap(A):
  size = len(A)
  for i in range((size - 2) // 2, -1, -1):
    heapifyTopDown(A, i)
  return

A = [1, 2, 3, 4, 5]
buildMaxHeap(A)
A
```

```
# insert an item to a max-heap
def insert(A, val):
  A.append(val)
  heapifyBottomUp(A, len(A) - 1)

def heapifyBottomUp(A, i):
  if i <= 0:
    return
  parent = (i - 1) // 2
  if A[parent] < A[i]:
    A[parent], A[i] = A[i], A[parent]
    heapifyBottomUp(A, parent)

A = [5, 4, 3, 1, 2]
insert(A, 6)
A
```

```
# pop the root item of a max-heap
def pop(A):
  if not A:
    raise Exception("heap is empty")
  A[0] = A[-1]
  del A[-1]
  heapifyTopDown(A, 0)

A = [6, 4, 5, 1, 2, 3]
pop(A)
A
```

```
# update an item in a max-heap
def update(A, i, val):
  if i < 0 or i >= len(A):
    raise Exception("index is out of bound of heap A")
  if val > A[i]:
    A[i] = val
    heapifyBottomUp(A, i)
  elif val < A[i]:
    A[i] = val
    heapifyTopDown(A, i)

A = [5, 4, 3, 1, 2]
update(A, 1, 0)
A
```

### Q1: Find smallest k elements from an unsorted array of size n

```
# method 1: build a max heap of size k
# for each of the remaining n-k elements, compare it to the top of the max heap
# if it's smaller, then popout the top item in the heap and push the current compared item to the heap
# The final heap is the returnd k smallest elements
# time complexity: O(k + (n - k)logk)
# space complexity: O(k)

import heapq
def kSmall(a, k):
    if not a or k >= len(a):
        return a
    # build k-size max heap
    b = [-a[i] for i in range(k)]
    heapq.heapify(b)
    for i in range(len(a) - k):
        if a[k + i] < -b[0]:
            heapq.heappop(b)
            heapq.heappush(b, -a[k + i])
    return [-item for item in b]

# test
a = [1,3,2,4,0]
k = 3
kSmall(a, k)
```

```
# method 2: build a min heap of size n
# pop from the min heap for k times
# these k items are the smallest k elements
# time complexity: O(n + k logn)
# space complexity: O(k)

import heapq
def kSmall(a, k):
    '''
    input:
        a: list of numbers
        k: integer
    output:
        list of k smallest elements in a
    '''
    if not a or k >= len(a):
        return a
    # build a min heap
    heapq.heapify(a)
    result = []
    for _ in range(k):
        result.append(heapq.heappop(a))
    return result

# test
a = [1,3,2,4,0]
k = 3
kSmall(a, k)
```

```
def partition(A):
  if not A:
    return -1
  i, j = 0, len(A) - 2
  pivot = A[-1]
  while j >= i:
    if A[j] < pivot:
      A[i], A[j] = A[j], A[i]
      i += 1
    elif A[j] > pivot:
      j -= 1
  # put pivot in the middle
  A[i], A[-1] = A[-1], A[i]
  return i

A = [1, 2, 6, 4, 5, 3] 
i = partition(A)
print(A, i)
```

```
# quick partition in-place
# quick partition
# choose a random number as pivot (usually the last element), partite the list into two parts
# left part is smaller than pivot
# right part is larger than pivot
# if length of left part is larger than k, then right part can be discarded
# if length of left part is smaller than k, then it can be added to the result, and find the remaining k - len(left) from the right part
# if length of left part is k, then it's the result
# time complexity: worst case O(n^2), average case, each time half of items can be discarded, O(n + n/2 + n/4 + ...) = O(n)

def kSmall(a, k, left, right, res):
    '''
    input:
        a: list of numbesr
        k: integer
    output: a list of k smallest item in the list a
    '''
    if not a:
        return
    if k == right - left + 1:
        res.extend(a[left:right + 1])
        return
    # base case
    if k == 0 or left > right:
        return
    # partite inplace
    idx = partite(a, left, right)
    if idx - left + 1 > k:
        kSmall(a, k, left, idx, res)
    elif idx - left + 1 < k:
        res.extend(a[left:idx + 1])
        kSmall(a, k - (idx - left + 1), idx + 1, right, res)
    else:
        res.extend(a[left:idx + 1])
        return
    
# use two pointers i and j
# take the right most item as the pivot
# left to i is the partite smaller than pivot
# right to j is the partite larger than pivot
# from i to j is the unknow part
def partite(a, left, right):
    '''
    input:
        a: a list of numbers
        [left, right]: partite range
    output:
        partite the list in-place, return the index of the middle number (pivot)
    '''
    pivot = a[right]
    i, j = left, right - 1
    while i <= j:
        if a[i] <= pivot:
            i += 1
        else:
            a[i], a[j] = a[j], a[i]
            j -= 1
    a[i], a[right] = a[right], a[i]
    return j


# test     
a = [5,4,3,2,1,0]
res = []
kSmall(a, 3, 0, len(a) - 1, res)
res
```

## Graph Search Algorithm: Breadth First Search (BFS)

### Breadth First Search (BFS-1)

```
# Graph node class
class GraphNode:
  def __init__(self, val = None):
    self.val = val
    self.neighbor = set()
  
  def add(self, nodeList):
    for node in nodeList:
      self.neighbor.add(node)

graphNode1 = GraphNode(1)
graphNode2 = GraphNode(2)
graphNode3 = GraphNode(3)
graphNode4 = GraphNode(4)

graphNode1.add([graphNode2, graphNode4])
graphNode2.add([graphNode1, graphNode3])
graphNode3.add([graphNode2, graphNode4])
graphNode4.add([graphNode1, graphNode3])
```

```
# BFS1
# initialize an queue, add the root node to a queue
# while queue is not empty
# pop out an element in queue, add the not yet visited neighbor of the popped out node into the queue
# use a visited set to check if a node is visited or not
# time complexity: O(n), n is the number of nodes in the tree
# space complexity: O(n), the max number of nodes in the queue can be n/2, so the space complexity is O(n)

from collections import deque
def BFS1(graphNode):
    q = deque([graphNode])
    visited = set([graphNode]) # or {graphNode}
    while q:
        node = q.popleft()
        print(node.val)
        for neighb in node.neighbor:
            if neighb not in visited:
                q.append(neighb)
                visited.add(neighb)
        
# test
BFS1(graphNode1)
```

### 经典例题2: Bipartite: whether a graph's node can be divided into two group, such that the nodes in each group do not have direct edges between the nodes that belong to the same group

```
def bipartite(graphNode):
  if not graphNode:
    return True
  u = set()
  v = set()
  queue1 = [graphNode]
  queue2 = []
  while queue1 or queue2:
    while queue1:
      for node in queue1[0].neighbor:
        if node in u:
          return False
        if node not in v:
          v.add(node)
          queue2.append(node)
      del queue1[0]

    while queue2:
      for node in queue2[0].neighbor:
        if node in v:
          return False
        if node not in u:
          u.add(node)
          queue1.append(node)
      del queue2[0]
  return True

bipartite(graphNode1) 
```

### 经典例题3: Determine whether a binary tree is a complete binary tree

```
from collections import deque
def isComplete(root):
  if not root:
    return True
  q = deque([root])
  hasChild = True
  while q:
    node = q.popleft()
    if hasChild:
      if node.left:
        q.append(node.left)
      else:
        hasChild = False
        if node.right:
          return False
    else: # hasChild == False
      if node.left or node.right:
        return False
  return True

isComplete(treeNode1)
```

### Best First Search (BFS-2)
Dijkstra's Algorithm

```
# Graph node with edge weight class
class GraphNodeW:
  def __init__(self, val = None):
    self.val = val
    self.neighbor = {}
  
  def add(self, pairList):
    for pair in pairList:
      self.neighbor[pair[0]] = pair[1]
  
  def __lt__(self, other):
    if self.val < other.val:
      return True
    else:
      return False

graphNodeW1 = GraphNodeW(1)
graphNodeW2 = GraphNodeW(2)
graphNodeW3 = GraphNodeW(3)
graphNodeW4 = GraphNodeW(4)

graphNodeW1.add([(graphNodeW2, 1), (graphNodeW4, 1)])
graphNodeW2.add([(graphNodeW1, 1), (graphNodeW3, 1)])
graphNodeW3.add([(graphNodeW2, 1), (graphNodeW4, 1)])
graphNodeW4.add([(graphNodeW1, 1), (graphNodeW3, 1)])
```

```
import heapq
def BFS2(graphNodeW):
  if not graphNodeW:
    return
  q = [(0, graphNodeW)]
  res = {graphNodeW : 0}
  while len(q) > 0:
    cost, nodePoped = heapq.heappop(q)
    if nodePoped in res and cost > res[nodePoped]:
      continue
    for node, weight in nodePoped.neighbor.items():
      if node in res and cost + weight >= res[node]:
        continue
      res[node] = cost + weight
      heapq.heappush(q, (cost + weight, node))
  for node in res:
    print("node ", node.val, "cost ", res[node], "   ")

BFS2(graphNodeW1)
```

### Given a matrix of size NxN, and for each row the elements are sorted in an ascending order. and for each column the elements are also sorted in an ascending order. How to find the k-th smallest element in it?

```
import heapq
def kSmallest(mat, k):
  n = len(mat)
  m = len(mat[0])
  if k <= 1:
    return mat[0][0]
  if k > n * m:
    raise Exception("there is less than k items in the matrix")
    
  visited = {(0, 0)}
  count = 0
  pq = [(mat[0][0], 0, 0)]
  while pq:
    poped, i, j = heapq.heappop(pq)
    count += 1
    if count == k:
      return poped
    if i + 1 < n and (i + 1, j) not in visited:
      heapq.heappush(pq, (mat[i + 1][j], i + 1, j))
      visited.add((i + 1, j))
    if j + 1 < m and (i, j + 1) not in visited:
      heapq.heappush(pq, (mat[i][j + 1], i, j + 1))
      visited.add((i, j + 1))
    

mat = [[1,2,3], [4,5,6], [7, 8, 9]]
kSmallest(mat, 6)
```
