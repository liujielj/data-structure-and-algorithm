### DFS经典例题1: print all subsets of a set S={'a', 'b', 'c'}

```
def subSet(s):
  res = []
  subSetHelper(list(s), 0, [], res)
  return res

def subSetHelper(sList, idx, aList, res):
  if idx == len(sList):
    # append a copy of aList into res, otherwise, the aList content appended will change later
    res.append(aList[:])
    return
  # Case 1: include the current item
  aList.append(sList[idx])
  subSetHelper(sList, idx + 1, aList, res)
  aList.pop()
  # Case 2: not include the current item
  subSetHelper(sList, idx + 1, aList, res)
  
  
s = {'a', 'b', 'c'}
subSet(s)
```

### DFS经典例题2: ()()() find all valid permutation using the parenthesis provided

```
def parenPermu(n):
  res = []
  helper(n, 0, 0, [], res)
  return res

def helper(n, l, r, permu, res):
  if len(permu) == 2 * n:
    res.append(permu[:])
    return
  # left "("
  if l < n:
    permu.append('(')
    helper(n, l + 1, r, permu, res)
    permu.pop()
  
  # right ")"
  if r < l:
    permu.append(')')
    helper(n, l, r + 1, permu, res)
    permu.pop()

parenPermu(3)
```

### DFS经典例题3：Print all combinations of coins that can sum up to total value k.
E.g. total value k = 99 cents, coin value = 25 10 5 1 cent

```
def coinComb(k, coins):
  res = []
  comb = [0] * len(coins)
  helper(k, coins, 0, comb, res)
  return res

def helper(k, coins, idx, comb, res):
  if k == 0:
    res.append(comb[:])
    return
  if idx == len(coins):
    return
  n = k // coins[idx]
  for i in range(n + 1):
    comb[idx] = i
    helper(k - i * coins[idx], coins, idx + 1, comb, res)
    #comb[idx] = 0


coinComb(12, [25, 10, 5, 1])
#coinComb(50, [50, 25, 10, 5, 1])
```

### DFS经典例题4：Given a string with no duplicated letters, how to print out all permutations of the string.

```
# Solution 1
def permu(s):
  res = []
  helper(s, [], res)
  return res

def helper(s, permuS, res):
  if not s:
    res.append(permuS[:])
    return
  #print(permuS, s)
  for i in range(len(s)):
    permuS.append(s[i])
    helper(s[:i] + s[i + 1 :], permuS, res)
    permuS.pop()

s = "abc"
permu(s)
```

```
# Solution 2
def permu(s):
  res = []
  used = [False] * len(s)
  helper(s, [], used, res)
  return res

def helper(s, permuS, used, res):
  if len(permuS) == len(s):
    res.append(''.join(permuS))
    return

  for i in range(len(s)):
    if used[i] == False:
      permuS.append(s[i]) 
      used[i] = True
      helper(s, permuS, used, res)
      used[i] = False
      permuS.pop()

s = "abc"
permu(s)
```

```
# Solution 3: swap-swap method
def permu(s):
  res = []
  helper(list(s), 0, res)
  return res

def helper(s, idx, res):
  if idx == len(s):
    res.append(''.join(s))
    return
  for i in range(idx, len(s)):
    s[i], s[idx] = s[idx], s[i]
    helper(s, idx + 1, res)
    s[i], s[idx] = s[idx], s[i]

s = "abc"
permu(s)
```
